### Step 4. Discover what is the mean age per occupation

```python
occupations = users['occupation'].unique()
print(occupations)

mean = np.zeros(21)
for i in range(21):
  occ = users[users['occupation'] == occupations[i]]
  mean[i] = occ['age'].mean()
mean_age_occ = pd.DataFrame({"occupation":occupations,
                   "mean_age":mean})
mean_age_occ
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

#### method1

```python
male_ratio = np.zeros(21)
for i in range(21):
  occ = users[users['occupation'] == occupations[i]]
  total = occ.shape[0] # number of people
  males = occ[occ['gender'] == 'M'].shape[0] # number of males
  male_ratio[i] = males/total
male_ratio_occ = pd.DataFrame({"occupation":occupations,
                   "male_ratio":male_ratio})
male_ratio_occ = male_ratio_occ.sort_values(by = ['male_ratio'], ascending=False)
print(male_ratio_occ)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

#### Method 1 (groupby)

```python
df_users = pd.DataFrame(users)

occ_age_min_max = df_users.groupby(by = 'occupation')['age'].agg([max,min])
print(occ_age_min_max)
```

#### Method 2

```python
min_age = np.zeros(21)
max_age = np.zeros(21)
for i in range(21):
  occ = users[users['occupation'] == occupations[i]]
  min_age[i] = occ['age'].min()
  max_age[i] = occ['age'].max()
age_minmax_occ = pd.DataFrame({"occupation":occupations,
                               "min_age":min_age.astype('int32'),
                               "max_age":max_age.astype('int32')})
age_minmax_occ
```

```python
occ_gender = list(np.repeat(occupations,2))
mean_age = np.zeros(42)
def repeat_list_flat(input_list, n):
    return input_list * n
gender = ['M','F']
gender = repeat_list_flat(gender, 21)

for i in range(21):
  occ = users[users['occupation'] == occupations[i]]
  occ_m = occ[occ['gender'] == 'M']
  occ_f = occ[occ['gender'] == 'F']
  mean_age[2*i] = occ_m['age'].mean()
  mean_age[2*i+1] = occ_f['age'].mean()

meanage_gender_occ = pd.DataFrame({"occupation":occ_gender,
                               "gender":gender,
                               "mean_age":mean_age.astype('int32')})
meanage_gender_occ
```

### Step 8.  For each occupation present the percentage of women and men

### Method1

```python
male_perc = np.zeros(21)
female_perc = np.zeros(21)
for i in range(21):
  occ = users[users['occupation'] == occupations[i]]
  total = occ.shape[0] # number of people
  males = occ[occ['gender'] == 'M'].shape[0] # number of males
  females = occ[occ['gender'] == 'F'].shape[0] # number of females
  male_perc[i] = (males/total)*100
  female_perc[i] = (females/total)*100
gender_perc_occ = pd.DataFrame({"occupation":occupations,
                                "male_perc":male_perc,
                                "female_perc":female_perc})
gender_perc_occ
```
