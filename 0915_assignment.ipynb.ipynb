### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv).

```python
data = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv")
```

### Step 3. Assign it to a variable called tips

```python
tips = data
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
tips = tips.drop(['Unnamed: 0'], axis=1)
tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
plt.hist(tips['total_bill'])
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
figure = plt.figure()
axes1 = figure.add_subplot(111)
axes1.scatter(tips['total_bill'],tips['tip'])
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
figure = plt.figure()
axes1 = figure.add_subplot(111)
x = tips['total_bill']
y = tips['tip']
s = tips['size'] * 10
axes1.scatter(x,y,s)
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### Step 8. Present the relationship between days and total_bill value

```python
tips['day'].value_counts()
```

```python
# days :  
# total_bill :  

fig = plt.figure()
axes1 = fig.add_subplot(111)
x1 = tips[tips['day']=='Sat']['total_bill']
x2 = tips[tips['day']=='Sun']['total_bill']
x3 = tips[tips['day']=='Thur']['total_bill']
x4 = tips[tips['day']=='Fri']['total_bill']
labels = ['Thur','Fri','Sat', 'Sun']

axes1.boxplot([x1,x2,x3,x4],labels=labels)
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
def recode_sex(sex):
    if sex=='Female':
        return 0
    else :
        return 1

tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
fig = plt.figure()
axes = fig.add_subplot(111)

x = tips['tip']
y = tips['day']
s = tips['size'] * 10
c = tips['sex_color']


axes.scatter(x, y, s, c)
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
print(tips['time'].value_counts())
print(tips['day'].value_counts())
```

```python
#   : day & time
#   : total_bill

fig = plt.figure()
axes = fig.add_subplot(111)
x1 = tips[tips['day']=='Thur'][tips['time']=='Lunch']['total_bill']
x2 = tips[tips['day']=='Thur'][tips['time']=='Dinner']['total_bill']
x3 = tips[tips['day']=='Fri'][tips['time']=='Lunch']['total_bill']
x4 = tips[tips['day']=='Fri'][tips['time']=='Dinner']['total_bill']
x5 = tips[tips['day']=='Sat'][tips['time']=='Lunch']['total_bill']
x6 = tips[tips['day']=='Sat'][tips['time']=='Dinner']['total_bill']
x7 = tips[tips['day']=='Sun'][tips['time']=='Lunch']['total_bill']
x8 = tips[tips['day']=='Sun'][tips['time']=='Dinner']['total_bill']
labels = ['thur lunch', 'thur dinner', 'Fri Lunch', 'Fri Dinner', 'Sat Lunch', 'Sat Dinner', 'Sun Lunch', 'Sun Dinner']

axes.boxplot([x1, x2, x3, x4, x5, x6, x7, x8], labels = labels)
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
fig = plt.figure()
axes1 = fig.add_subplot(121)
axes2 = fig.add_subplot(122)
x1 = tips[tips['time']=='Lunch']['tip']
x2 = tips[tips['time']=='Dinner']['tip']

axes1.hist(x1)
axes2.hist(x2)

axes1.set_title('Lunch Tip')
axes1.set_xlabel('Tip')
axes1.set_ylabel('Density')

axes2.set_title('Dinner Tip')
axes2.set_xlabel('Tip')
axes2.set_ylabel('Density')
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
tips.head()
```

```python
def recode_smoking(smoker):
    if smoker=='No':
        return 'r'
    else :
        return 'b'

tips['smoke_color'] = tips['smoker'].apply(recode_smoking)
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(121)
axes2 = fig.add_subplot(122)


x1 = tips[tips['sex']=='Female']['total_bill']
y1 = tips[tips['sex']=='Female']['tip']

x2 = tips[tips['sex']=='Male']['total_bill']
y2 = tips[tips['sex']=='Male']['tip']

c1 = tips[tips['sex']=='Female']['smoke_color']
c2 = tips[tips['sex']=='Male']['smoke_color']



axes1.scatter(x=x1, y=y1, color=c1)
axes2.scatter(x=x2, y=y2, color=c2)

axes1.set_title('Famale')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

axes2.set_title('Male')
axes2.set_xlabel('Total Bill')
axes2.set_ylabel('Tip')
```

### Create two boxplot graphs, one for Male and another for Female, presenting the total_bill values , differing by day

```python
fig = plt.figure()
axes1 = fig.add_subplot(121)
axes2 = fig.add_subplot(122)

x1 = tips[tips['day']=='Thur'][tips['sex']=='Female']['total_bill']
x2 = tips[tips['day']=='Fri'][tips['sex']=='Female']['total_bill']
x3 = tips[tips['day']=='Sat'][tips['sex']=='Female']['total_bill']
x4 = tips[tips['day']=='Sun'][tips['sex']=='Female']['total_bill']
x5 = tips[tips['day']=='Thur'][tips['sex']=='Male']['total_bill']
x6 = tips[tips['day']=='Fri'][tips['sex']=='Male']['total_bill']
x7 = tips[tips['day']=='Sat'][tips['sex']=='Male']['total_bill']
x8 = tips[tips['day']=='Sun'][tips['sex']=='Male']['total_bill']

labels = ['Thur', 'Fri', 'Sat', 'Sun']

axes1.boxplot([x1, x2, x3, x4], labels = labels)
axes2.boxplot([x5, x6, x7, x8], labels = labels)

axes1.set_title('Famale')
axes1.set_ylabel('Total Bill')
axes1.set_xlabel('Day')

axes2.set_title('Male')
axes2.set_ylabel('Total Bill')
axes2.set_xlabel('Day')
```
